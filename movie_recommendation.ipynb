##### Converting genres columns into right format

```python
def change(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres']= df['genres'].apply(change)
```

```python
df
```

```python
df['keywords'] = df['keywords'].apply(change)
```

```python
def convert(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l   
    
```

```python
df['cast'] = df['cast'].apply(convert)
```

```python
df
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
        
```

```python
df['crew'] = df['crew'].apply(director)
```

```python
df['overview'][0]
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df
```


#### Remove space between two words because each word may be similar so model may confused during recommendation so by combining the two words, each word will be unique.


```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['keywords'] = df['keywords'].apply(lambda x: [i.replace(" ",'') for i in x])
df['cast'] = df['cast'].apply(lambda x: [i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.drop(['overview','genres','keywords','cast','crew'],axis = 1,  inplace = True)
```

#### Use of Stemming to convert similar words into unique or same words

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
        
        
```
